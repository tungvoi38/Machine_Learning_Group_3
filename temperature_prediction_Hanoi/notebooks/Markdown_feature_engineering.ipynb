# 1. Biến dẫn xuất cơ bản từ dữ liệu thời tiết

Biến ***spread*** được tính bằng hiệu giữa nhiệt độ cao nhất và thấp nhất trong ngày (tempmax - tempmin), phản ánh biên độ dao động nhiệt độ, giúp mô hình nhận biết sự thay đổi mạnh yếu của khí hậu trong ngày.


***temp_dew_diff*** là chênh lệch giữa nhiệt độ và điểm sương (temp - dew), cho biết mức độ ẩm tương đối của không khí – khi chênh lệch nhỏ, độ ẩm cao, khả năng hình thành sương hoặc mưa tăng.

***feelslike_diff*** biểu thị độ lệch giữa nhiệt cảm nhận và nhiệt thực (feelslike - temp), giúp mô hình hiểu được ảnh hưởng của gió và độ ẩm đến cảm nhận nhiệt.

***is_heatwave*** là biến nhị phân nhận giá trị 1 nếu ba ngày trước đó đều có nhiệt độ cao nhất trên 35°C, phản ánh chuỗi nắng nóng kéo dài, và được tính bằng cách sử dụng rolling window kết hợp shift để tránh rò rỉ dữ liệu.

***PET (Potential Evapotranspiration)*** ước lượng khả năng bốc hơi nước qua công thức kết hợp năng lượng mặt trời và chênh lệch nhiệt độ, cung cấp thông tin về điều kiện khô hạn.

***daylight_duration_hours*** đo thời lượng ban ngày (tính từ lúc mặt trời mọc đến lặn), biểu thị tổng thời gian nhận năng lượng mặt trời trong ngày.

Hai biến ***wind_U*** và ***wind_V*** chuyển hướng gió (đơn vị độ) thành hai thành phần tọa độ sin và cos của vector gió, giúp mô hình học hướng gió một cách liên tục mà không gặp vấn đề tuần hoàn (0° và 360°).

***pressure_daily_change*** là mức thay đổi áp suất trong ba ngày trước, phản ánh xu hướng thời tiết (áp thấp, áp cao) có thể ảnh hưởng đến nhiệt độ.

***solar_cloud_interaction*** kết hợp bức xạ mặt trời và tỷ lệ mây che phủ để mô phỏng lượng năng lượng thực tế đến mặt đất.

***wind_gust_ratio*** là tỉ lệ giữa gió giật và gió trung bình, dùng để phát hiện những thời điểm gió mạnh bất thường so với trạng thái bình thường.

# 2. Biến thời gian (Datetime decomposition)

Nhóm này mô tả đặc tính tuần hoàn của thời gian. Các biến ***month_sin*** và ***month_cos*** mã hóa chu kỳ 12 tháng, giúp mô hình nhận diện mùa nóng, mùa mưa.

Tương tự, ***day_sin*** và ***day_cos*** mô tả vị trí ngày trong năm, nắm được sự thay đổi nhiệt độ theo mùa.

Hai biến ***dow_sin*** và ***dow_cos*** thể hiện chu kỳ tuần, cho phép mô hình nhận ra khác biệt giữa ngày thường và cuối tuần.

***week_sin*** và ***week_cos*** mô phỏng chu kỳ theo tuần ISO (52 tuần), phù hợp với các mô hình nhận dạng dài hạn.

Biến ***quarter*** biểu thị quý (1–4), phản ánh mùa hoặc giai đoạn trong năm.

Cuối cùng, ***is_weekend*** là cờ nhận giá trị 1 cho thứ 7, Chủ nhật — đặc biệt hữu ích nếu mô hình phân tích dữ liệu hành vi hoặc hoạt động phụ thuộc thời gian.

# 3. Biến dẫn xuất nâng cao từ điều kiện thời tiết

***heat_index*** kết hợp nhiệt độ và điểm sương để ước lượng cảm giác nóng ẩm, thể hiện tác động của độ ẩm đến cảm nhận nhiệt.

***humidity_ratio*** đo tỷ lệ độ ẩm hiện tại so với độ ẩm cao nhất trong 7 ngày trước, cho biết không khí hôm nay ẩm hay khô hơn trung bình gần đây.

Hai biến ***pressure_trend_1d*** và ***pressure_trend_3d*** thể hiện xu hướng thay đổi áp suất trong 1 và 3 ngày, giúp mô hình nhận biết chuyển động khí quyển.


# 4. Biến tương tác giữa các yếu tố

***temp_humidity_interaction*** là tích giữa nhiệt độ và độ ẩm, phản ánh tương tác phi tuyến — khi nhiệt cao và ẩm cao cùng lúc, ảnh hưởng đến cảm nhận nhiệt mạnh hơn.

***temp_wind_interaction*** là tích giữa nhiệt độ và tốc độ gió, thể hiện hiệu ứng làm mát do gió.

***temp_cloud_interaction*** mô tả sự ảnh hưởng của mây lên khả năng hấp thụ nhiệt; khi mây dày, năng lượng mặt trời giảm và nhiệt độ thay đổi chậm hơn.


# 5. Biến cờ thời tiết đặc biệt

***is_coldwave*** nhận giá trị 1 nếu hai ngày trước đó có nhiệt độ tối thiểu không vượt quá 10°C, phản ánh giai đoạn lạnh kéo dài.

***is_rainy*** bằng 1 khi có lượng mưa dương, biểu thị ngày mưa.

***is_high_wind*** bằng 1 nếu tốc độ gió lớn hơn phân vị 90%, đánh dấu hiện tượng gió mạnh.

# 6. Biến chuyển đổi phân phối và phi tuyến

***precip_log*** và ***solar_log*** được tính bằng logarit tự nhiên của lượng mưa và bức xạ mặt trời (sau khi clip giá trị âm), giúp làm mượt các giá trị cực đoan và giảm skewness.

***temp_squared*** và ***humidity_squared*** là bình phương của nhiệt độ và độ ẩm, cho phép mô hình học mối quan hệ phi tuyến giữa biến này và mục tiêu.

# 7. Biến rolling thống kê theo thời gian

Các biến ***rolling*** được tính cho nhiều cột thời tiết như độ ẩm, lượng mưa, tốc độ gió, áp suất, bức xạ mặt trời, độ phủ mây… với 4 khung thời gian: 7, 28, 56 và 91 ngày.
Với mỗi khung thời gian, bốn thống kê được tính gồm trung bình, phương sai, giá trị nhỏ nhất và lớn nhất, tất cả đều được shift(1) để chỉ sử dụng dữ liệu quá khứ.

Ví dụ: 7D_AVG_HUMIDITY là độ ẩm trung bình 7 ngày trước, 56D_VAR_TEMP là phương sai nhiệt độ trong 56 ngày trước.
Những biến này giúp mô hình hiểu xu hướng, độ ổn định và mức dao động khí hậu ở các khung thời gian khác nhau.


# 8. Biến lag (giá trị trễ)

Các biến lag được tạo cho nhiều chỉ số thời tiết như nhiệt độ, độ ẩm, áp suất, lượng mưa, hướng gió, v.v. với độ trễ 1, 3, 5 và 7 ngày.
        
Ví dụ, TEMP_LAG_1 là nhiệt độ của ngày hôm trước, HUMIDITY_LAG_7 là độ ẩm của 7 ngày trước.
Nhóm này mô tả cấu trúc tự hồi quy trong dữ liệu thời tiết — tức là giá trị hiện tại có thể phụ thuộc vào quá khứ gần.


# 9. Biến mục tiêu (Targets)

Cuối cùng, năm biến mục tiêu. ***y_temp_1*** đến. ***y_temp_5*** được tạo bằng cách dịch ngược cột nhiệt độ 1–5 ngày, lần lượt đại diện cho nhiệt độ thực tế trong 1, 2, 3, 4 và 5 ngày tương lai.
Nhóm này phục vụ cho bài toán multi-step forecasting — dự đoán nhiệt độ cho nhiều ngày liên tiếp.


# 10. Tổng kết

Tổng cộng, quá trình feature engineering tạo ra hơn 500 đặc trưng, bao gồm các biến gốc, biến phi tuyến, tương tác, rolling, và lag. Tất cả đều được thiết kế để phản ánh hành vi vật lý của thời tiết và tránh rò rỉ dữ liệu bằng cách sử dụng shift(1) trước khi tính toán.

Những đặc trưng này cung cấp cho mô hình học máy nhiều thông tin hơn về xu hướng, tính chu kỳ, và tương tác khí hậu — giúp cải thiện độ chính xác khi dự báo nhiệt độ trong tương lai.
