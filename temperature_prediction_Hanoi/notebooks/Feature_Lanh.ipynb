```python
import pandas as pd
import numpy as np
```

```python
file_path = '/Users/user/Desktop/HN/Hanoi Daily.csv'
df = pd.read_csv(file_path)
```

```python
# Kiểm tra kiểu dữ liệu từng cột
print(df.dtypes)
```

```python
# Danh sách các biến được phép tạo rolling như bạn cung cấp
features = [
    "tempmax", "tempmin", "temp", "feelslikemax", "feelslikemin", "feelslike",
    "dew", "humidity", "precip", "precipcover", "windgust", "windspeed",
    "winddir", "sealevelpressure", "cloudcover", "visibility", "solarradiation",
    "solarenergy", "uvindex", "severerisk", "sunrise"
]

# Biến nhiệt độ không rolling dài mà chỉ rolling 7 ngày
temp_vars = ["tempmax", "tempmin", "temp", "feelslikemax", "feelslikemin", "feelslike"]
```

```python

# Chỉ chọn column có kiểu số để tính rolling (một số biến có thể là object hoặc category)
numeric_vars = df.select_dtypes(include=[np.number]).columns.tolist()

# Lọc biến trong features mà ở dạng số
numerical_features = [f for f in features if f in numeric_vars]

# Biến nhiệt độ thuộc danh sách số
temp_vars_num = [v for v in temp_vars if v in numerical_features]

# Các biến còn lại cần rolling dài
vars_to_roll = [f for f in numerical_features if f not in temp_vars_num]

windows = list(range(7, 92, 7))  # 7,14,...,91

df_features = df.copy()

# Rolling mean 7 ngày cho biến nhiệt độ
for var in temp_vars_num:
    df_features[f'7D_AVG_{var.upper()}'] = df[var].rolling(window=7, min_periods=1).mean()

# Rolling mean và variance cho các biến còn lại với các window
for var in vars_to_roll:
    for w in windows:
        df_features[f'{w}D_AVG_{var.upper()}'] = df[var].rolling(window=w, min_periods=1).mean()
        df_features[f'{w}D_VAR_{var.upper()}'] = df[var].rolling(window=w, min_periods=1).var()


# df_features.to_csv('/Users/user/Desktop/HN/Hanoi_Daily_Features.csv', index=False)
```
